**Step-10: Data Visualisation**

```python
plt.figure(figsize=(20,10))
plt.bar('Sub-Category','Category', data=data)
plt.title('Category vs Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=50)
plt.show()
```

```python
data.hist(figsize=(16,16))
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.countplot(x=data['State'])
plt.xticks(rotation=50)
plt.title('States')
plt.show()
```

```python
corr = data.corr()
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
plt.figure(figsize=(10,5))
a = sns.heatmap(corr, annot=True, fmt='.1f', cmap='YlGnBu')
rotx = a.set_xticklabels(a.get_xticklabels(), rotation=90)
roty = a.set_yticklabels(a.get_yticklabels(), rotation=30)
```

```python
Profit_plot=(ggplot(data, aes(x='Sub-Category', y='Profit', fill='Sub-Category')) + geom_col() + coord_flip()+ scale_fill_brewer(type='div', pelette='Spectral') + theme_classic() + ggtitle('pie chart'))
display(Profit_plot)
```

```python
ggplot(data, aes(x='Ship Mode', fill ='Category')) + geom_bar(stat ='count')
```

```python
flip_xlabels = theme(axis_text_x = element_text(angle=90, hjust=1),figure_size=(10,5),
                    axis_ticks_length_major=10,axis_ticks_length_minor=5)
(ggplot(data, aes(x='Sub-Category', fill='Sales'))+ geom_bar()+ facet_wrap(['Segment'])+flip_xlabels + theme(axis_text_x = element_text(size=13))+ggtitle("Sales Every parts of USA"))
```

```python
plt.figure(figsize=(10,8))
data['Category'].value_counts().plot(kind='pie',autopct='%.2f')
plt.title('Category')
plt.ylabel('')
plt.show()
```

```python
plt.style.use("seaborn")
fig, ax = plt.subplots(figsize=(8,8))
plt.pie(x=data["Ship Mode"].value_counts(), 
        colors=["skyblue","pink",'green','red'], 
        labels=["Standard Class","Second Class",'First Class','Same Day'], 
        shadow = True, 
        autopct="%1.2f%%", 
        explode = (0, 0.1,0.2,0.2)
        )
plt.show()
```
